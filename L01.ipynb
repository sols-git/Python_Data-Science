1. Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

**Ответ:**

*Precision* - точность отражает какой процент объектов, отнесённых к "основному" классу, действительно являются таковыми.

*Recall* - полнота отражает какой процент объектов "основного" класса было обнаружено алгоритмом.

$$\large precision = \LARGE \frac{TP}{TP \: + \: FP} \:\:\:\:\: \large recall = \LARGE \frac{TP}{TP \: + \: FN}$$

Предположим, на производстве поставлена задача спрогнозировать момент во времени (или периодичность до износа), когда производственный тот или иной агрегат выйдет из строя.
Критичность прогноза сбоев агрегатов должна соответствовать экономической целесообразности при определении периодов их технического обуслуживания. Например у нас есть 2 класса агрегатов: А - критичный, то есть при поломке таких агрегатов прелприятие остановит выпуск продукции и Б - не критиный, то есть при его сбое основная функция предприятия не потеряет в качестве и не оставновит производство.
Соответственно, при неточности прогноза определения периодов сбоя для категории А (критичного), с недостаточной **Precision**, экономический ущерб предприятия может на порядки превышать соответствующие инвестиии для ремонта или замены на исправное оборудования в рамках предупредительного технического обслуживания. И наоборот, если прогноз момента поломки некритичного оборудования Б, будет неточным с недостаточной **Recall** , то предприятие будет затрачивать необоснованные экономические усилия в рамках предупредительного ремонта.

Поэтому для агрегатов типа А (критичные) важно максимизировать **Precision**,
а для агрегатов Б (некритичные) важно максимизировать **Recall**

2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

**Ответ:**

F-score* - метрика объединяющая в себе точность и полноту, выдавая как результат их гармоническое среднее, а коэффициент $\beta$, позволяет балансирвоать вклад, вносимый точностью, тем самым акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна.

$$\LARGE F_{\beta} = (1+ \beta^2) \cdot \frac{precision \: \cdot \: recall}{(\beta^2 \cdot precision) \: + \: recall}$$

Мы не можем взять средние значения от Precision и Recall постольку поскольку мы не сможем управлять требуемым видом точности определения ошибок первого и второго рода в зависимости от целей задачи. А F мера дает такие возможности.
