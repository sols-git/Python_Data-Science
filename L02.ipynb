1. В чём различие между зависимыми и независимыми выборками?

Ответ: зависимая выборка предполагает некоторую заисимость выбираемых параметров из таких выборок между собой, а незаивисимые выборки 
       не имеют данной связи между их значениями, и вероятно имеют разные распределения


2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Ответ: параметрические статистики применяются для нормальный распределений, непараметрические могут применяются для неномальных распределений
